```python
path = 'C:/Users/ethan/Downloads/datagame-2023'
train = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/label_train_source.parquet')
test = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/label_test_source.parquet')
composer = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song_composer.parquet')
genre = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song_genre.parquet')
lyricist = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song_lyricist.parquet')
producer = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song_producer.parquet')
titletext = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song_titletext.parquet')
song = pd.read_parquet('C:/Users/ethan/Downloads/datagame-2023/meta_song.parquet')
```

# merge data


```python
song_data = pd.merge(song, composer, on='song_id', how='left')
del composer, song
song_data = pd.merge(song_data, genre, on='song_id', how='left')
del genre
song_data = pd.merge(song_data, lyricist, on='song_id', how='left')
del lyricist
song_data = pd.merge(song_data, producer, on='song_id', how='left')
del producer
song_data = pd.merge(song_data, titletext, on='song_id', how='left')
del titletext
msno.matrix(song_data)
```

# 新增區段

```python
for i in song_data.select_dtypes(include=['object']).columns:
    song_data[i][song_data[i].isnull()] = 'unknown'
song_data = song_data.fillna(value=0)
msno.matrix(song_data)
```

```python
label_encoder = LabelEncoder()
one_hot = OneHotEncoder()

for i in song_data.columns :
    song_data[i] = label_encoder.fit_transform(song_data[i])
song_data.head()
```
